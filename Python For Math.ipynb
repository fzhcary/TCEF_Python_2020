```python
max_step = [0, 0]

def do_magic(i):
    global max_step
    i_save = i
    step = 0
    print(i, end='')
    while i != 1:
        if i % 2 == 0:
            i = i//2
        else:
            i = i * 3 + 1
        step += 1
        print(f'->{i}', end='')
    if step > max_step[0]:
        max_step = [step, i_save]
    print(f"... it took {step} steps to come to 1")
    

for i in range(1, 100):
    do_magic(i)

print(max_step)
```

```python
import math
print(2, end=" ")
for i in range(3, 100):
    is_prime = True
    if i % 2 == 0:
        continue
        
    for j in range(3, int(math.sqrt(i)), 2):
        if i % j == 0:
            is_prime = False
            break
            
    if is_prime:
        print(i, end=" ")
```

```python
from functools import reduce

# def check_duplicate(l):
#     return len(set(l))<=1
    
def get_num_from_list(l):
    return reduce(lambda x, y: 10*x+y, l)

def get_digits_from_number(number):
    return list(map(lambda x:int(x), str(number)))

def search_number(args, search=0, depth=1):
#     print("search_number", args, depth)
    number = get_num_from_list(args)
    print(number, end=" > ")
    
    small = get_num_from_list(sorted(args))
    big = get_num_from_list(sorted(args, reverse=True))
    
    result = big - small
    if result == search:
#         print(f"{result}...it took {depth} steps")
        return depth
    elif depth < 100:
        digits = get_digits_from_number(result)
        return search_number(digits, result, depth+1)
    else:
#         print("not found")
        return None
```

```python
result = []
for i in range(10, 10000):
    result.append([i, search_number(get_digits_from_number(i))])
```

```python
def check(n):
    x = 0
    y = 0
    for i in range(1,n):
        x += i**3
        y += i
    if x != y**2:
        return False
    else:
        return True
    
for i in range(10):
    print(check(i))
```
